```python
import hh
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint 
```

```python
eqns = hh.HodgkinHuxley()
```

```python
Y = np.array([0.0, eqns.n_inf(), eqns.m_inf(), eqns.h_inf()])
Vy = odeint(eqns.derivatives, Y, eqns.t)
V = Vy[:,0]
m = Vy[:,1]
h = Vy[:,2]
n = Vy[:,3]

plt.figure()

plt.subplot(4,1,1)
plt.title('Hodgkin-Huxley for space-clamped axon')
plt.plot(eqns.t, V, 'k')
plt.ylabel('V (mV)')
plt.xlabel('Time (ms)')

plt.subplot(4,1,3)
plt.plot(eqns.t, m, 'r', label='m')
plt.plot(eqns.t, h, 'g', label='h')
plt.plot(eqns.t, n, 'b', label='n')
plt.ylabel('Gating Value')
plt.xlabel('Time (ms)')
plt.legend()

plt.show()
```
